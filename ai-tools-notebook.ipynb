```python
copilot = data[data['AI_Tools_Used'] == 'Copilot']
use_case_counts = copilot['Use_Cases'].value_counts().head(5)
from collections import Counter
use_cases_series = copilot['Use_Cases'].dropna()

all_use_cases = []
for entry in use_cases_series:
    all_use_cases.extend([case.strip() for case in entry.split(',')])
use_case_counts = pd.Series(Counter(all_use_cases)).sort_values(ascending=False).head(5)
plt.figure(figsize=(10,6))
use_case_counts.plot(kind='bar', color='royalblue')

plt.title("Top 5 AI Use Cases for Copilot", fontsize=14)
plt.xlabel("Use Case")
plt.ylabel("Number of Students")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
chatgpt= data[data['AI_Tools_Used'] == 'ChatGPT']
use_case_counts = chatgpt['Use_Cases'].value_counts().head(5)
from collections import Counter
use_cases_series = chatgpt['Use_Cases'].dropna()

all_use_cases = []
for entry in use_cases_series:
    all_use_cases.extend([case.strip() for case in entry.split(',')])
use_case_counts = pd.Series(Counter(all_use_cases)).sort_values(ascending=False).head(5)
plt.figure(figsize=(10,6))
use_case_counts.plot(kind='bar', color='royalblue')

plt.title("Top 5 AI Use Cases for chat gpt", fontsize=14)
plt.xlabel("Use Case")
plt.ylabel("Number of Students")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
